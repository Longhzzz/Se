```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../../../data/stf/'
n3,n2,n1 = 150,200,1000
```

```python
sx = np.fromfile(path+'sx_t_flatten.dat',dtype=np.float32).reshape(n3,n2,n1)
fc = np.fromfile(path+'fc_t_flatten.dat',dtype=np.float32).reshape(n3,n2,n1)
ai = np.fromfile(path+'acoustic_impedance_t_flatten.dat',dtype=np.float32).reshape(n3,n2,n1)
```

```python
plt.subplot(131)
plt.imshow(sx[100,:,:980].T,cmap='seismic')
plt.subplot(132)
plt.imshow(ai[100,:,:980].T,cmap='seismic')
plt.subplot(133)
plt.imshow(fc[100,:,:980].T,cmap='seismic')
```
